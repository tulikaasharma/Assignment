```python
cols =['fname','mname','lname']
dfpk["full_name"] = dfpk[cols].apply(lambda x: ' '.join(x.dropna()), axis=1)
```

```python
dfpk['_1st_absent_on'] = dfpk["fname"].map(
    pd.melt(dfpk, id_vars="fname")
    .query('value == "."')
    .groupby("fname")["variable"].first()
)
```

```python
dfpk['_last_absent_on'] = dfpk["fname"].map(
    pd.melt(dfpk, id_vars="fname")
    .query('value == "."')
    .groupby("fname")["variable"].last()
)
dfpk
```

```python
conditions =[
    (dfpk['_1st_absent_on'] == dfpk['_last_absent_on']),
    (pd.isna(dfpk['_1st_absent_on']) and pd.isna(dfpk['_last_absent_on'])),
    (dfpk['_1st_absent_on'] != dfpk['_last_absent_on'])]
     
values = ['oh!!','wow','shame-shame']
dfpk['bolo_kab_kab']= np.select(conditions, values)
```
